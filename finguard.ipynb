# FinGuard IntelliAgent - Kaggle Capstone Project

**Objective:** Build a multi-agent financial assistant for Kenyan SMEs using the Google Agent Development Kit (ADK).

**Tools:**
* Python
* `google-agent-development-kit`
* `google-generativeai` (for Gemini)

---

### Project Plan

**Milestone 1: Foundation & PoC**
* **Goal:** Create the orchestrator (`FinGuardIntelliAgent`) and one LLM-powered sub-agent (`SmsRoutingAgent`).
* **Tech:** `agents.Agent`, `llm_client`, `asyncio`.
* **Kaggle Concepts:** Multi-agent system, LLM-powered agent.

**Milestone 2: Core Sub-Agent & Custom Tool**
* **Goal:** Build the `PaymentAgent` and a custom `PaymentTool`.
* **Tech:** `tools.Tool`, `tool_client`, `llm_client.generate_content_with_tools`.
* **Kaggle Concepts:** Custom tools.

**Milestone 3: Integration, Memory & State**
* **Goal:** Add `AccountBalanceAgent` and implement session memory.
* **Tech:** `sessions.InMemorySessionService`.
* **Kaggle Concepts:** Sessions & Memory.

**Milestone 4: Final Packaging & Write-up**
* **Goal:** Consolidate code and write the Kaggle project description.
* **Tech:** Markdown.
